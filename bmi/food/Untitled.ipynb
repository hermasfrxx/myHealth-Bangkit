```python
import pandas as pd
```

```python
data = pd.read_csv('nutrients_csvfile.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
columns_to_drop = ['Fiber','Sat.Fat', 'Carbs', 'Measure']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.info()
```

```python
data = data[data['Fat'] != "t'"]
```

```python
unique_values = {
    'Food': data['Food'].unique(),
    'Grams': data['Grams'].unique(),
    'Calories': data['Calories'].unique(),
    'Protein': data['Protein'].unique(),
    'Fat':data['Fat'].unique(),
    'Category': data['Category'].unique()
}

for column, values in unique_values.items():
    print(f"Unique values for {column}:")
    print(values)
```

```python
data = data[data['Protein'] != "t'"]
```

```python
data = data[data['Calories'] != "nan'"] 
```

```python
data.shape
```

```python
data = data[data['Category'] != "Drinks,Alcohol, Beverages''"] 
```

```python
data['Protein'].unique()
```

```python
# Mengkonversi nilai 't' menjadi NaN dan mengubah tipe kolom menjadi numerik jika belum
data['Protein'] = pd.to_numeric(data['Protein'], errors='coerce')

# Menghapus baris yang memiliki nilai NaN pada kolom 'Protein'
data = data.dropna(subset=['Protein'])

# Menampilkan nilai unik pada kolom 'Protein'
print(data['Protein'].unique())
```

```python
data = data[data['Category'] != 'Drinks,Alcohol, Beverages']

# Menampilkan kategori unik setelah menghilangkan kategori tertentu
print(data['Category'].unique())
```

```python
data.shape
```

```python
data.info()
```

```python
# Mengkonversi kolom 'Grams' dan 'Calories' menjadi float
data['Grams'] = pd.to_numeric(data['Grams'], errors='coerce').astype(float)
data['Calories'] = pd.to_numeric(data['Calories'], errors='coerce').astype(float)
```

```python
data.describe()
```

```python
data = data.dropna(subset=['Grams', 'Calories', 'Protein'])
```

```python
data.describe()
```

```python
data['Protein'].unique()
```

```python
data = data[data['Protein'] != -1]
```

```python
data['Protein'].unique()
```

```python
# Mengatur ulang batas bin untuk 3 kategori
bins = [data['Calories'].min(), data['Calories'].quantile(0.33), data['Calories'].quantile(0.66), data['Calories'].max()]
labels = ['Rendah', 'Sedang', 'Tinggi']

# Menambahkan kolom baru dengan kategori berdasarkan nilai BMR
data['Calories_Category'] = pd.cut(data['Calories'], bins=bins, labels=labels)
# Inisialisasi LabelEncoder
```

```python
data.info()
```

```python
data.drop(columns=['Calorie_Level'])
```

```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
label_encoder = LabelEncoder()

# Melakukan encoding pada kolom BMR_Category
data['Id_Calories'] = label_encoder.fit_transform(data['Calories_Category'])
```

```python
data.info()
```

```python
data.drop(columns=['Caloriee_Category_Encoded'])
```

```python
data.to_csv('calorie_data.csv',index=False)
```
